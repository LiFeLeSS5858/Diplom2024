Поиск Координат

```python
from geopy.geocoders import Nominatim
nominatim = Nominatim(user_agent='user')
for i in range(len(df1)):
  place = str(df1.iloc[i,18])+" "+str(df1.iloc[i,19])
  nominatim = Nominatim(user_agent='user')
  try:
    df1.iloc[i,20] = "["+str(nominatim.geocode(place).latitude)+", "+str(nominatim.geocode(place).longitude)+"]"
  except AttributeError:
    place = str(df1.iloc[i,17])
    nominatim = Nominatim(user_agent='user')
    try:
      df1.iloc[i,20] = "["+str(nominatim.geocode(place).latitude)+", "+str(nominatim.geocode(place).longitude)+"]"
    except AttributeError:
      df1.iloc[i,20] = str(0)
```

Поиск Страны

```python
from geopy.geocoders import Nominatim

def get_location_info(latitude, longitude):
    geolocator = Nominatim(user_agent="geo_locator")
    location = geolocator.reverse((latitude, longitude), language="ru")

    if location:
        address = location.raw
        country = address.get('address', {}).get('country', 'N/A')
        state = address.get('address', {}).get('state', 'N/A')
        city = address.get('address', {}).get('town', 'N/A')

        return country, state, city
    else:
        return 'N/A', 'N/A', 'N/A'

for i in range(len(df1)):
  s = df1.iloc[i,20]
  s = s.replace('[','').replace(']','')
  x, y = s.split(', ')
  x = float(x)
  y = float(y)
  latitude = x
  longitude = y
  country, state, city = get_location_info(latitude, longitude)
  df1.iloc[i,17] = country
  #df1.iloc[i,18] = state
  #df1.iloc[i,19] = city
```

Изменяем поля направлений (большой шрифт, убрать ненужные символы)

```python
import re
c = '-'
for i in range(len(df1)):
  for j in range(4):
    try:
      df1.iloc[i,j+2]=df1.iloc[i,j+2].upper()
      df1.iloc[i,j+2] = re.sub(r'\(.*\)', '', df1.iloc[i,j+2])
      df1.iloc[i,j+2] = re.findall('.*' + c, df1.iloc[i,j+2])[0][:-1]
    except IndexError:
      df1.iloc[i,j+2] = df1.iloc[i,j+2]
    except TypeError:
      df1.iloc[i,j+2] = df1.iloc[i,j+2]
    except AttributeError:
      df1.iloc[i,j+2] = df1.iloc[i,j+2]
```
